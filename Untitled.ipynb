```python
import pandas as pd
import re



def standardize_semester(sem):
    sem = str(sem).lower()

    # Extract numeric values
    sem_numeric = re.search(r'\d+', sem)
    if sem_numeric:
        sem_numeric = sem_numeric.group()
    else:
        sem_numeric = ""

    # Map common representations to standardized format
    if '1st' in sem or 'first' in sem or 'fresher' in sem:
        return '1st'
    elif '2nd' in sem or 'second' in sem:
        return '2nd'
    elif '3rd' in sem or 'third' in sem:
        return '3rd'
    elif '4th' in sem or 'fourth' in sem:
        return '4th'
    elif '5th' in sem or 'fifth' in sem:
        return '5th'
    elif '6th' in sem or 'sixth' in sem:
        return '6th'
    elif '7th' in sem or 'seventh' in sem:
        return '7th'
    elif '8th' in sem or 'eighth' in sem:
        return '8th'
    elif '9th' in sem or 'ninth' in sem:
        return '9th'
    elif '10th' in sem or 'tenth' in sem:
        return '10th'
    elif '11th' in sem or 'eleventh' in sem:
        return '11th'
    elif '12th' in sem or 'twelfth' in sem:
        return '12th'
    elif '13th' in sem or 'thirteenth' in sem:
        return '13th'
    elif '14th' in sem or 'fourteenth' in sem:
        return '14th'
    elif '15th' in sem or 'fifteenth' in sem:
        return '15th'
    elif 'final' in sem:
        return 'final'
    else:
        return sem_numeric

# Apply the function to the 'Curr_Sem' column
df['Curr_Sem1'] = df['Curr_Sem'].apply(standardize_semester)

```

```python
import pandas as pd

def preprocess_sm_hours(df):
    """Preprocesses the 'SM_Hours' column to convert values into numeric hours.

    Args:
        df: A pandas DataFrame containing the 'SM_Hours' column.

    Returns:
        A pandas DataFrame with the preprocessed 'SM_Hours' column.
    """

    df['SM_Hours'] = df['SM_Hours'].str.replace(r'\D', '', regex=True)  # Remove non-numeric characters

    def handle_imprecise_values(value):
        """Converts imprecise values to numerical ranges."""
        if pd.isna(value) or value == '':
            return pd.NA  # Return NA for empty strings or NaN values
        try:
            value = float(value)  # Convert to float if the value is already numeric
        except (ValueError, TypeError):
            pass  # Continue with the original value if conversion fails

        if isinstance(value, float):
            return value  # Return float values as-is

        if '-4' in str(value):
            return 3.5  # Midpoint of 3-4 hours
        elif '-5' in str(value):
            return 4.5  # Midpoint of 4-5 hours
        elif '-6' in str(value):
            return 5.5  # Midpoint of 5-6 hours
        elif '-7' in str(value):
            return 6.5  # Midpoint of 6-7 hours
        elif '/' in str(value):
            return float(str(value).replace('/', '.'))  # Convert fractions to decimals
        elif 'more than' in str(value).lower():
            return float(str(value).lower().replace('more than ', ''))  # Retain value as a lower bound
        elif 'less than' in str(value).lower():
            return 0  # Assume 0 hours for "less than" responses
        else:
            return float(value)

    df['SM_Hours1'] = df['SM_Hours'].apply(handle_imprecise_values).astype(float)

    # Handle additional cases and potential errors:
    df.loc[df['SM_Hours1'] > 24, 'SM_Hours1'] = 24  # Cap values at 24 hours
    df.loc[df['SM_Hours1'] < 0, 'SM_Hours1'] = pd.NA  # Set negative values to NA
    df.loc[df['SM_Hours1'].isna(), 'SM_Hours1'] = pd.NA  # Fill missing values with NA

    return df

# Apply the preprocessing function:
# df1 = preprocess_sm_hours(df1.copy())  # Work on a copy to avoid modifying the original DataFrame
```

```python
import pandas as pd
import re

# Assuming df1 is your DataFrame
# Convert the column to string to handle non-string entries
df1['Avg_Attendance'] = df1['Avg_Attendance'].astype(str)

# Define a function to clean and standardize attendance values
def clean_attendance(value):
    # Remove non-alphanumeric characters
    value = re.sub(r'[^a-zA-Z0-9%]', '', value)
    
    # Extract percentage values
    if '%' in value:
        # Extract numeric values only
        value = re.search(r'\d+', value).group()
        value = int(value)
        # Standardize percentages to be within 0-100 range
        value = min(max(value, 0), 100)
    else:
        # If not a percentage, try to convert to integer
        try:
            value = int(value)
        except ValueError:
            # If conversion fails, set to NaN
            value = float('nan')

        # Ensure the value is within 0-100 range
        value = min(max(value, 0), 100)

    return value

# Apply the clean_attendance function to the 'Avg_Attendance' column
df1['Avg_Attendance1'] = df1['Avg_Attendance'].apply(clean_attendance)

# Drop rows with NaN values if necessary
# df1 = df1.dropna(subset=['Avg_Attendance'])

# Print unique values after preprocessing
print(df1['Avg_Attendance1'].unique())
```
